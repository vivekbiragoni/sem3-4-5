```python
import numpy as np
from qiskit.visualization import plot_bloch_vector
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_qsphere(qubit_states):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    for state in qubit_states:
        bloch_coords = [np.real(state[1]), np.imag(state[1]), np.real(state[0])]
        plot_bloch_vector(bloch_coords, ax=ax)

    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_zlim([-1, 1])
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('QSphere')

    plt.show()

# Example usage
# qubit_states = [(0.5, 0.5), (0.5j, -0.5j)]
qubit_states = [(0.5, 0.5), (0.5j, -0.5j), (0.707, 0)]

plot_qsphere(qubit_states)
```

```python
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_state_city

# Create a multi-qubit circuit
circuit = QuantumCircuit(2)
circuit.h(0)
circuit.cx(0, 1)

# Simulate the circuit and obtain the density matrix
backend = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend).result()
density_matrix = result.get_statevector()

# Visualize the density matrix
plot_state_city(density_matrix)
```
